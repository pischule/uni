### Legend

![](https://i.ibb.co/GQHqLDc/3.png)

```
# Plot two or more lines on same plot with suitable legends of each line
plt.plot([10, 20, 30], [20, 40, 10], color='blue', label='line 1')
plt.plot([10, 20, 30], [40, 10, 30], color='green', label='line 2')
plt.title('Two lines on same plot with legends')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
pass
```

### Color & Width

![](https://i.ibb.co/mvk1vvF/4.png)

```
# Plot two or more lines with legends, different widths and colors
plt.plot([10, 20, 30], [20, 40, 10], '-b', label='line1-width-3', linewidth=3)
plt.plot([10, 20, 30], [40, 10, 30], '-r', label='line2-width-5', linewidth=5)
plt.legend()
pass
```

![](https://i.ibb.co/8z6gjGP/5.png)

### Three lines - one plt.plot()

![](https://i.ibb.co/NTr8CJF/9.png)

```
# Plot several lines with different format styles in one command
import numpy as np
x = np.linspace(1, 20, 20)
ynlog = x * np.log(x)
ysquare = np.power(x, 2)

plt.plot(x, x, 'g--', x, ynlog, 'bs', x, ysquare, 'r^')
pass
```

### Major grid

![](https://i.ibb.co/VCYfbJd/10.png)

```
# Display the major grid and draw lines of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
# Customize the grid lines with linestyle -, width 0.5 and blue color

import matplotlib.dates as mdates

date_fmt = mdates.DateFormatter("%Y-%m-%d")
days = mdates.DayLocator()

xdates = [date(2016, 10, i) for i in range(3, 8)]
ydata = [772.5, 776.4, 776.5, 776.7, 775.0]

fig, ax = plt.subplots()
ax.plot(xdates, ydata, color="red", marker="o")
ax.grid(True, color="blue", ls="-", linewidth=.5)

ax.xaxis.set_major_formatter(date_fmt)
ax.xaxis.set_major_locator(days)

ax.yaxis.set_major_locator(plt.MultipleLocator(.5))

pass
```

### Minor grid

![](https://i.ibb.co/YffFjrJ/11.png)

```
# Display the magor and minor grids and draw lines of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
# Customize the grid lines with a larger grid (major grid) and a smaller grid (minor grid).
# Turn on the grid but turn off ticks

import matplotlib.dates as mdates
from matplotlib.dates import rrule, RRuleLocator, AutoDateLocator
from matplotlib.dates import HOURLY

date_fmt = mdates.DateFormatter("%Y-%m-%d")
days = mdates.DayLocator()
hours = mdates.HourLocator()

xdates = [date(2016, 10, i) for i in range(3, 8)]
ydata = [772.5, 776.4, 776.5, 776.7, 775.0]

fig, ax = plt.subplots()
ax.plot(xdates, ydata, color="red", marker="o")

ax.grid(True, which="major", color="red", ls="-", linewidth=.5)
ax.xaxis.set_major_formatter(date_fmt)
ax.xaxis.set_major_locator(days)
ax.yaxis.set_major_locator(plt.MultipleLocator(.5))

ax.grid(True, which="minor", color="black", ls="--", linewidth=.5)
ax.yaxis.set_minor_locator(plt.MultipleLocator(.1))
loc = AutoDateLocator()
loc.intervald[HOURLY] = [6]
ax.xaxis.set_minor_locator(loc)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

pass
```
